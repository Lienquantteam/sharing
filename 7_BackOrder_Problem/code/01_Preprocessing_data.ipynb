# Handling repetitive values:

It seems there are many  0's in our dataset. Let's find out the **repetition of 0's **in each column. If any of the columns has more than 60%, 0 we can safely drop that column. This kind of data trend can be viewed on violinplot. 

</br>



```python
fig , ax1 = plt.subplots(ncols=2)
sns.violinplot(data.Forecast_sales_3_months,   ax=ax1[0])
#plt.title('transit_quantity range and distribution')

sns.violinplot(data.Transit_time,  ax=ax1[1])
plt.title('Transit_time range and distribution')
```

```python
# calculate the repeatition of 0 in each column
perct_0  = {column: '%2.2f' % ( ((data[column] == 0).sum()/(data.shape[0]))*100 ) + 
            '%' for column in data.columns if (data[column].dtype == np.float64) or 
            data[column].dtype == np.int}
```

```python
perct_0 
```

```python
del_cols = [k for k, v in perct_0.items() if float(v.split("%")[0]) > 60]
```

```python
del_cols
```

```python
list(data.columns)
```

```python
# drop the column having more than 60% of 0's
data = data.drop(del_cols, axis=1)
```

Lets define a funtion to find the whisker values (in other words, upper and lower IQR boundaries for outliers

```python
def get_outlier_boundaries(column):
    q75, q25  = np.percentile(column, [75,25])
    iqr = q75 - q25

    lower = q25 - 1.5 * iqr
    upper = q75 + 1.5 * iqr

    return lower,upper
```

```python
transit_duration_upper = get_outlier_boundaries(data.Transit_time)[1]
outlier_transit_duration = data[data.Transit_time > transit_duration_upper]
print(outlier_transit_duration.shape)
outlier_transit_duration.Backordered.value_counts()
```
