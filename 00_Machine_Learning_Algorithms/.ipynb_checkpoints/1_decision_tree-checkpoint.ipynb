```python
from sklearn import tree
import pandas as pd
```

```python
dataset = pd.read_csv('dataset/1_mushrooms.csv', sep=",")
dataset.info()
```

```python
dataset.shape
```

```python
dataset.head(10)
```

#### normalization

```python
dataset['cap-shape'] = dataset['cap-shape'].map({'b': 1, 
                                                 'c': 2,
                                                 'x': 3,
                                                 'f': 4,
                                                 'k' : 5,
                                                 's':6 })

dataset['cap-surface'] = dataset['cap-surface'].map({'f': 1,
                                                     'g': 2,
                                                     'y': 3,
                                                     's': 4 })

dataset['cap-color'] = dataset['cap-color'].map({'n': 1,
                                                 'b': 2,
                                                 'c': 3,
                                                 'g': 4,
                                                 'r' : 5,
                                                 'p':6,
                                                 'u':7,
                                                 'e':8, 
                                                 'w':9,
                                                 'y':10 })

dataset['bruises'] = dataset['bruises'].map({'t': 1, 
                                             'f': 0 })

dataset['odor'] = dataset['odor'].map({'a': 1,
                                       'l': 2,
                                       'c': 3,
                                       'y': 4,
                                       'f' : 5,
                                       'm': 6,
                                       'n' : 7,
                                       'p': 8,
                                       's': 9})

dataset['gill-attachment'] = dataset['gill-attachment'].map({'a': 1, 
                                                             'd': 2,
                                                             'f': 3,
                                                             'n': 4 })

dataset['gill-spacing'] = dataset['gill-spacing'].map({'c': 1,
                                                       'w': 2,
                                                       'd': 3 })

dataset['gill-size'] = dataset['gill-size'].map({'b': 1, 
                                                 'n': 2 })

dataset['gill-color'] = dataset['gill-color'].map({'k': 1,
                                                   'n': 2,
                                                   'b': 3,
                                                   'h': 4,
                                                   'g' : 5,
                                                   'r':6,
                                                   'o':7,
                                                   'p':8,
                                                   'u':9,
                                                   'e':10,
                                                   'w' : 11,
                                                   'y' : 12 })

dataset['stalk-shape'] = dataset['stalk-shape'].map({'e': 1,
                                                     't': 2 })

dataset['stalk-root'] = dataset['stalk-root'].map({'b': 1,
                                                   'c': 2,
                                                   'u': 3, 
                                                   'e': 4,
                                                   'z' : 5,
                                                   'r' : 6,
                                                   '?': 0 })

dataset['stalk-surface-above-ring'] = dataset['stalk-surface-above-ring'].map({'f': 1,
                                                                               'y': 2,
                                                                               'k': 3,
                                                                               's': 4})

dataset['stalk-surface-below-ring'] = dataset['stalk-surface-below-ring'].map({'f': 1,
                                                                               'y': 2,
                                                                               'k': 3,
                                                                               's': 4})

dataset['stalk-color-above-ring'] = dataset['stalk-color-above-ring'].map({'n': 1,
                                                                           'b': 2,
                                                                           'c': 3,
                                                                           'g': 4,
                                                                           'o' : 5,
                                                                           'p': 6,
                                                                           'e' : 7,
                                                                           'w':8,
                                                                           'y':9})

dataset['stalk-color-below-ring'] = dataset['stalk-color-below-ring'].map({'n': 1,
                                                                           'b': 2,
                                                                           'c': 3, 
                                                                           'g': 4,
                                                                           'o' : 5,
                                                                           'p': 6,
                                                                           'e' : 7,
                                                                           'w':8,
                                                                           'y':9})

dataset['veil-type'] = dataset['veil-type'].map({'p': 1, 
                                                 'u': 2 })

dataset['veil-color'] = dataset['veil-color'].map({'n': 1,
                                                   'o': 2,
                                                   'w':3,
                                                   'y':4 })

dataset['ring-number'] = dataset['ring-number'].map({'n': 0,
                                                     'o': 1,
                                                     't': 2 })

dataset['ring-type'] = dataset['ring-type'].map({'c': 1,
                                                 'e': 2,
                                                 'f':3, 
                                                 'l':4, 
                                                 'n':0,
                                                 'p':5, 
                                                 's':6,
                                                 'z': 7})

dataset['spore-print-color'] = dataset['spore-print-color'].map({'k': 1, 
                                                                 'n': 2,
                                                                 'b':3, 
                                                                 'h':4, 
                                                                 'r':5,
                                                                 'o':6,
                                                                 'u':7, 
                                                                 'w':8,
                                                                 'y':9 })

dataset['population'] = dataset['population'].map({'a': 1,
                                                   'c': 2, 
                                                   'n':3, 
                                                   's':4,
                                                   'v':5,
                                                   'y':6 })

dataset['habitat'] = dataset['habitat'].map({'g': 1, 
                                             'l': 2, 
                                             'm':3, 
                                             'p':4,
                                             'u':5,
                                             'w':6,
                                             'd':7 })

dataset['class'] = dataset['class'].map({'p': 1,
                                         'e': 2 })
```

```python
dataset.head(10)
```

```python
dataset.head(10)
```

```python
dataset.info()
```

split into train and test sets

```python
X = dataset.drop(['class'], axis=1)
y = dataset['class']
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) 
```

```python
tree_n = tree.DecisionTreeClassifier()
tree_n.fit(X_train,y_train)
```

predict in test set

```python
prediction = tree_n.predict(X_test)
prediction
```

accuracy

```python
print("The prediction accuracy is: ",tree_n.score(X_test,y_test)*100,"%")
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test, prediction))  
print(classification_report(y_test, prediction))  
```

```python
# write to txt file
with open("mushroom.txt", "w") as f:
    f = tree.export_graphviz(tree_n, out_file=f)
```
